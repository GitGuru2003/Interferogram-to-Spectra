```python
import os
from brukeropus import read_opus

DATA_DIR = "../Data"

def is_opus_file(fname: str) -> bool:
    # Accept numeric extensions: .0 .1 .0001 etc.
    _, ext = os.path.splitext(fname)
    return len(ext) > 1 and ext[1:].isdigit()

results = {}

for fname in sorted(os.listdir(DATA_DIR)):
    if not is_opus_file(fname):
        continue

    path = os.path.join(DATA_DIR, fname)
    if not os.path.isfile(path):
        continue

    try:
        op = read_opus(path)

        try:
            phz = op.params["phz"]     # preferred
        except Exception:
            # some files/libraries use uppercase keys
            phz = op.params["PHZ"]

        results[fname] = phz
        print(f"{fname}  -->  PHZ = {phz}")

    except Exception as e:
        print(f"{fname}  -->  ERROR reading file: {e}")

print("\n=== Summary ===")
counts = {}
for v in results.values():
    counts[v] = counts.get(v, 0) + 1
for mode, c in sorted(counts.items(), key=lambda x: (-x[1], x[0])):
    print(f"{mode}: {c} files")
```
